**Data Description** - This dataset contains information about Indian School Dropout Statistics of the year 2013-2014 to 2015-2016. 


```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the file.
df=pd.read_csv("C:\\Users\\Anisha Siwas\\Downloads\\3-dropout-ratio-2012-2015.csv")
```

```python
# Head function will tell you the top 5 records of the dataset. By default python shows the top 5 records.
df.head()
```

```python
# Describe is used to view some basic statistical details like, percentile, mean, standard deviation, etc of the given dataframe.
df.describe()
```

```python
# Info tells about how many null and non-null values are there in the dataset and what are dtypes of each column.
df.info()
```

```python
# Changing dtypes from object to variables
df['Primary_Total'] = pd.to_numeric(df['Primary_Total'], errors='coerce')
df['Primary_Boys'] = pd.to_numeric(df['Primary_Boys'], errors='coerce')
df['Primary_Girls'] = pd.to_numeric(df['Primary_Girls'], errors='coerce')
df['Upper Primary_Boys'] = pd.to_numeric(df['Upper Primary_Boys'], errors='coerce')
df['Upper Primary_Girls'] = pd.to_numeric(df['Upper Primary_Girls'], errors='coerce')
df['Upper Primary_Total'] = pd.to_numeric(df['Upper Primary_Total'], errors='coerce')
df['Secondary _Boys'] = pd.to_numeric(df['Secondary _Boys'], errors='coerce')
df['Secondary _Girls'] = pd.to_numeric(df['Secondary _Girls'], errors='coerce')
df['Secondary _Total'] = pd.to_numeric(df['Secondary _Total'], errors='coerce')
df['HrSecondary_Boys'] = pd.to_numeric(df['HrSecondary_Boys'], errors='coerce')
df['HrSecondary_Girls'] = pd.to_numeric(df['HrSecondary_Girls'], errors='coerce')
df['HrSecondary_Total'] = pd.to_numeric(df['HrSecondary_Total'], errors='coerce')
```

```python
df.dtypes
```

```python
#replacing null values with 0
df=df.replace('NR',0)
df=df.replace('Uppe_r_Primary',0)
df=df.replace('@',0)
```

```python
# Head function will tell you the top 5 records of the dataset. By default python shows the top 5 records.
df.head()
```

```python
# Info is used to check the information about the data and the datatypes of each respective attribute.
df.info()
```

```python
plt.figure(figsize=(40,10))
sns.boxplot(x='State_UT',y='Primary_Total',data=df)
# Primary total is highest for Mizoram state, the median is the highest.
```

```python
plt.figure(figsize=(40,10))
sns.boxplot(x='State_UT',y='Secondary _Total',data=df)
# Secondary total is highest for Odisha state, the median is the highest.
```

```python
plt.figure(figsize=(40,10))
sns.boxplot(x='State_UT',y='HrSecondary_Total',data=df)
```

```python
sns.lineplot(x='year',y='HrSecondary_Total', estimator=np.mean, hue='State_UT', data=df)
# In 2014-15 huge increase in dropout ratio in Dadea & Nagar Haveli.
```

```python
sns.lineplot(x='year',y='Secondary _Total', estimator=np.mean, hue='State_UT', data=df)
#In initial years highest drop out ratio in Odisha state.
```

```python
ax=pd.crosstab(df.State_UT,df.year)
```

```python
sns.set_context()
sns.violinplot(x='year',y='HrSecondary_Total',data=df)
```

```python
sns.violinplot(x='year',y='HrSecondary_Total',data=df)
sns.swarmplot(x='year',y='HrSecondary_Total',data=df,color='w')
```

```python
dp_hm=df.copy() # copy for heat map plotting
dp_hm.drop(['year','Primary_Total','Upper Primary_Total','Secondary _Total','HrSecondary_Total'],axis=1,inplace=True)
dp_hm.drop(index=dp_hm[dp_hm['State_UT']=='AllIndia'].index,inplace=True)
```

```python
dp_hm=dp_hm.groupby('State_UT').mean()
plt.style.use('fivethirtyeight')

ax=plt.subplots(figsize=(20,10))
ax=sns.heatmap(dp_hm,linewidths=0.02,annot=True)
plt.title('Heat Map')
plt.show()
```

# Conclusion

1. It can concluded from the dataset that Primary dropout ratio is highest for the Mizoram state, Secondary dropout ratio is highest for the Odisha state.
2. In 2014-15 huge increase in dropout ratio in Dadea & Nagar Haveli.
3. In initial years highest drop out ratio in Odisha state.
4. The mean dropout ratio for HrSecondary > Primary.
5. Higher drop out ratio in Tripura in 2013-14 and 2014-15 which was lower in 2012-13.
6. Overall dropout ratio is highest for Secondary boys and girls as compared with everything else in all the states. This dataset's conclusion can be used to decrease the dropout ratio targeting the Secondary Boys and Girls on priority basis while targeting other groups as well.
